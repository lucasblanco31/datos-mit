```python
import pandas as pd
import numpy as np
import glob
import os
import time
```

```python
key=pd.read_csv("tc4tl_dev_key.tsv" , sep='\t')
```

```python
todos=[]
contador=0
file=key['fileid']
pcs=key['phone_carriage_state']
dim=key['distance_in_meters']
ssis=key['step_size_in_sec']
cg=key['coarse_grain']
for f in glob.glob("*.csv"):
    df= pd.read_csv(f ,header=None, sep='\n')
    df= df[0].str.split(',', expand=True)
    txdunion=np.array(df.iloc[0,1])
    txpunion=np.array(df.iloc[1,1])
    rxdunion=np.array(df.iloc[2,1])
    txcarryunion=np.array(df.iloc[3,1])
    rxcarryunion=np.array(df.iloc[4,1])
    rxposeunion=np.array(df.iloc[5,1])
    txposeunion=np.array(df.iloc[6,1])
    
    
    name=np.array(f)
    for k in range(len(file)):
        if f==file[k]:
            pcsunion=np.array(pcs[k])
            dimunion=np.array(dim[k])
            ssisunion=np.array(ssis[k])
            cgunion=np.array(cg[k])
    
    df2=df.drop([0,1,2,3,4,5,6])
    df2['archivo'] = name
    df2['pcsunion']= pcsunion
    df2['TXDevice'] =txdunion
    df2['TXcarry'] =rxcarryunion
    df2['TXpose'] =txposeunion
    df2['TXPower'] =txpunion
    df2['RXDivice'] =rxdunion
    df2['RXcarry'] =rxcarryunion
    df2['RXpose'] =rxposeunion
    df2['distance_in_meters']= dimunion
    df2['step_size_in_sec'] = ssisunion
    df2['coarse_grain'] = cgunion
    
    todos.append(df2)
    contador+=1

union= pd.concat(todos, ignore_index='True')
union.columns=["time", "sensor", "rssi/x", "y","z","heading4","heading5","heading6","file","phone_carriage_state", "TXDevice", "TXcarry","TXpose" , "TXPower", "RXDivice", "RXcarry","RXpose", "distance_in_meters","step_size_in_sec","coarse_grain"]
union
```

```python
union.to_csv("union/unionv03.csv",sep = ',', index=False)
```
