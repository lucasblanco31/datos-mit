```python
accelerometer = db[(db.sensor=='Accelerometer')]
accelerometer = accelerometer.drop(columns=['heading4', 'heading5', 'heading6'])
accelerometer.rename(columns={'rssi/x': 'ax', 'y': 'ay', 'z': 'az'}, inplace=True)
accelerometer.columns
```

```python
gyroscope = db[(db.sensor=='Gyroscope')]
gyroscope = gyroscope.drop(columns=['heading4', 'heading5', 'heading6'])
gyroscope.rename(columns={'rssi/x': 'gx', 'y': 'gy', 'z': 'gz'}, inplace=True)
gyroscope.columns
```

```python
attitude = db[(db.sensor=='Attitude')]
attitude = attitude.drop(columns=['heading4', 'heading5', 'heading6'])
attitude.rename(columns={'rssi/x': 'pitch', 'y': 'roll', 'z': 'yaw'}, inplace=True)
attitude.columns
```

```python
gravity = db[(db.sensor=='Gravity')]
gravity = gravity.drop(columns=['heading4', 'heading5', 'heading6'])
gravity.rename(columns={'rssi/x': 'x', 'y': 'y', 'z': 'z'}, inplace=True)
gravity.columns
```

```python
magnetic = db[(db.sensor=='Magnetic-field')]
magnetic = magnetic.drop(columns=['heading5', 'heading6'])
magnetic.rename(columns={'rssi/x': 'x', 'y': 'y', 'z': 'z', 'heading4' : 'accuracy'}, inplace=True)
magnetic.columns
```

```python
altitude = db[(db.sensor=='Altitude')]
altitude = altitude.drop(columns=['z','heading4','heading5', 'heading6'])
altitude.rename(columns={'rssi/x': 'x', 'y': 'y'}, inplace=True)
altitude.columns
```

```python
activity = db[(db.sensor=='Activity')]
activity = activity.drop(columns=['heading4','heading5', 'heading6'])
activity.rename(columns={'rssi/x': 'start_time', 'y': 'activity_type', 'z': 'confidence_level'}, inplace=True)
activity.columns
```

```python
heading = db[(db.sensor=='Heading')]
heading.rename(columns={'rssi/x': 'true_heading', 'y': 'magnetic_heading', 'z': 'heading_accuracy'
                       , 'heading4': 'x', 'heading5': 'y', 'heading6': 'z'}, inplace=True)
heading.columns
```

```python
bluetooth = db[(db.sensor=='Bluetooth')]
bluetooth = bluetooth.drop(columns=['y','z','heading4','heading5', 'heading6'])
bluetooth.rename(columns={'rssi/x': 'rssi', 'y': 'activity_type', 'z': 'confidence_level'}, inplace=True)
bluetooth.columns
```
